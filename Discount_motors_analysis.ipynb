```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
cars_sales= pd.read_csv(r"C:\Users\primo\Downloads\toyota (1).csv")
cars_sales.head()
```

```python
cars_sales.describe()
cars_sales.info()
```

```python
model_proportions = cars_sales['model'].value_counts(normalize=True)
models = list(model_proportions.index)
cars_sales['model'] = cars_sales['model'].astype(pd.CategoricalDtype(categories=models))
transmission_proportions = cars_sales['transmission'].value_counts(normalize=True)
transmissions = list(transmission_proportions.index)
cars_sales['transmission'] = cars_sales['transmission'].astype(pd.CategoricalDtype(categories=transmissions))
fueltype_proportions = cars_sales['fuelType'].value_counts(normalize=True)
fueltype = list(fueltype_proportions.index)
cars_sales['fuelType'] = cars_sales['fuelType'].astype(pd.CategoricalDtype(categories=fueltype))
```

```python
fig, ax= plt.subplots(1, 3, figsize=(15, 5))
ax[0].pie(model_proportions, labels=model_proportions.index, autopct='%1.1f%%')
ax[1].pie(transmission_proportions, labels= transmission_proportions.index, autopct='%1.1f%%')
ax[2].pie(fueltype_proportions, labels=fueltype_proportions.index, autopct='%1.1f%%')
ax[0].set_title("Distribucin por modelo")
ax[1].set_title("Distribucin por transmisin")
ax[2].set_title("Distribucin por combustible")
plt.show()
```

```python
sns.set_style('whitegrid')
sns.catplot(data=cars_sales, x='fuelType', kind='count')
plt.show()
```

```python
# Crear la figura y subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # 1 fila, 3 columnas

# Grfico 1: Distribucin de mpg por fuelType
sns.boxplot(data=cars_sales, x='fuelType', y='mpg', ax=axes[0], palette='tab10')
axes[0].set_title("Distribucin de mpg por fuelType")
axes[0].set_xlabel("Tipo de Combustible")
axes[0].set_ylabel("Millas por Galn (mpg)")

# Grfico 2: Distribucin de price por fuelType
sns.boxplot(data=cars_sales, x='fuelType', y='price', ax=axes[1], palette='tab10')
axes[1].set_title("Distribucin de price por fuelType")
axes[1].set_xlabel("Tipo de Combustible")
axes[1].set_ylabel("Precio (USD)")

# Grfico 3: Distribucin de tax por fuelType
sns.boxplot(data=cars_sales, x='fuelType', y='tax', ax=axes[2], palette='tab10')
axes[2].set_title("Distribucin de tax por fuelType")
axes[2].set_xlabel("Tipo de Combustible")
axes[2].set_ylabel("Impuesto")

# Ajustar diseo
plt.tight_layout()
plt.show()
```

```python
# Crear la figura y subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # 1 fila, 3 columnas

# Grfico 1: Histograma de mpg por fuelType
sns.histplot(data=cars_sales, x='mpg', hue='fuelType', multiple='stack', ax=axes[0], palette='tab10', kde=True)
axes[0].set_title("Distribucin de mpg por fuelType")
axes[0].set_xlabel("Millas por Galn (mpg)")
axes[0].set_ylabel("Frecuencia")

# Grfico 2: Histograma de price por fuelType
sns.histplot(data=cars_sales, x='price', hue='fuelType', multiple='stack', ax=axes[1], palette='tab10', kde=True)
axes[1].set_title("Distribucin de price por fuelType")
axes[1].set_xlabel("Precio (USD)")
axes[1].set_ylabel("Frecuencia")

# Grfico 3: Histograma de tax por fuelType
sns.histplot(data=cars_sales, x='tax', hue='fuelType', multiple='stack', ax=axes[2], palette='tab10', kde=True)
axes[2].set_title("Distribucin de tax por fuelType")
axes[2].set_xlabel("Impuesto")
axes[2].set_ylabel("Frecuencia")

# Ajustar diseo
plt.tight_layout()
plt.show()
```

```python
avg_price_by_fuel = cars_sales.groupby('fuelType')['price'].mean()
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_price_by_fuel.index, y=avg_price_by_fuel.values, palette='tab10')
plt.title("Precio Promedio por Tipo de Combustible")
plt.xlabel("Tipo de Combustible")
plt.ylabel("Precio Promedio (USD)")
plt.show()
```

```python
hybrids = cars_sales[cars_sales['fuelType'] == 'Hybrid']
hybrids['model'] = hybrids['model'].astype('object')
model_counts = hybrids['model'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=model_counts.index, y=model_counts.values, palette='tab10')
plt.title("Cantidad de Registros por Modelo (Solo Autos Hbridos)")
plt.xlabel("Modelo")
plt.ylabel("Cantidad de Registros")
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas para mejor visualizacin
plt.show()
```

```python
sns.histplot(data=cars_sales, x='price', bins=30)
plt.show()
```

```python
cars_sales_subset = cars_sales[cars_sales['fuelType'] != 'Other']
cars_sales_subset['fuelType'] = cars_sales_subset['fuelType'].astype('object')
g = sns.FacetGrid(cars_sales_subset, col="fuelType", col_wrap=3, sharex=True, sharey=True)
g.map(sns.histplot, "price", color="steelblue", bins=20)
plt.show()
```

```python
sns.boxplot(data=hybrids, x='model', y='price')
plt.show()
```

```python
tax_grouped = cars_sales_subset.groupby(['fuelType', 'tax']).agg(count=('model','count')).reset_index()
tax_pivot = tax_grouped.pivot(index='tax', columns='fuelType', values='count').fillna(0)
colors = sns.color_palette("tab10", n_colors=tax_pivot.shape[1])
ax = tax_pivot.plot(
    kind='bar', 
    stacked=True, 
    figsize=(12, 6), 
    color=colors
)
y_ticks = [0, 50, 100, 250, 500, 750, 1000, 1500, 2000]
plt.yticks(y_ticks)
ax.set_yticks(y_ticks)
plt.title('Number of Used cars by tax and Fuel type')
plt.xlabel('Tax')
plt.ylabel('Cantidad de Autos')
plt.xticks(rotation=45)
plt.legend(title='Fuel type', loc='upper right')
plt.tight_layout()

# Mostrar el grfico
plt.show()
```
